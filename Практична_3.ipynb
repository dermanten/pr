```python
#5

from math import sqrt

n = int(input("Введіть ціле число вкладень в корінь: "))

temp = sqrt(2)
for i in range(1, n):
  temp = sqrt(temp + 2)

print(temp)
```

```python
#7
n=int(input('ВВедіть натуральне число: '))

tmp= n
count=0
while tmp !=0:
  if(tmp % 10)% 2 == 0:
    count +=1
  tmp //=10
print(f'Кількість парних розрядів числа {n}: {count}')
```
