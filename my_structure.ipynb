```python
#3
class Student:
    def __init__(self, pip, dob, address, course, group):
        self.pip = pip
        self.dob = dob
        self.address = address
        self.course = course
        self.group = group

students = [
    Student("Стаськова Валентина Анатоліївна", "2007-02-11", "Вінниця", 1, "1-СОМ"),
    Student("Сичук Ангеліна Олександрівна", "2006-11-15", "Вінниця", 1, "1-СОМ"),
    Student("Кучерук Христина Петрівна", "2005-04-10", "Вінниця", 2, "2-СОМ"),
]

print("Студенти 2-го курсу:")
for student in students:
    if student.course == 2:
        print(student.pip)
```

```python
#4`
cars = [
    {"марка": "Toyota", "рік випуску": 2020, "ціна": 38000},
    {"марка": "BMW", "рік випуску": 2019, "ціна": 45000},
    {"марка": "Renault", "рік випуску": 2018, "ціна": 32000},
]

print("Усі наявні автомобілі на складі:")
for car in cars:
    print(car)

print("\nАвтомобілі з ціною менше ніж 40000 грн:")
for car in cars:
    if car["ціна"] < 40000:
        print(f"Марка: {car['марка']}, Рік: {car['рік випуску']}")
```

```python
#4
class Car:
    def __init__(self, brand, year, price):
        self.brand = brand
        self.year = year
        self.price = price

cars = [
    Car("Toyota", 2020, 38000),
    Car("BMW", 2019, 45000),
    Car("Renault", 2018, 32000),
]

print("Усі наявні автомобілі на складі:")
for car in cars:
    print(f"Марка: {car.brand}, Рік: {car.year}, Ціна: {car.price}")

print("\nАвтомобілі з ціною менше ніж 40000 грн:")
for car in cars:
    if car.price < 40000:
        print(f"Марка: {car.brand}, Рік: {car.year}")
```
