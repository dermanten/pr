**Список (list)** - це структура даних для зберігання об'єктів різних типів.  В інших мовах програмування подібна структура даних часто називається **масивом**. Список дуже схожий на масив, тільки в ньому можна зберігати об'єкти різних типів. Розмір списку не статичний, його можна змінювати.
Список  є змінним типом даних. Змінна, яка визначається як список, містить посилання на структуру в пам'яті, яка в свою чергу зберігає посилання на будь-які інші об'єкти або структури.



Основні операції зі списками:

Створення списку: my_list = [1, 2, 3]

Доступ до елементів: my_list[0]

Додавання елемента: my_list.append(4)

Вставка елемента: my_list.insert(1, 5)

Видалення елемента: my_list.remove(2)

Видалення за індексом: del my_list[0]

Перебір списку: for item in my_list:

Сортування: my_list.sort()

Реверс списку: my_list.reverse()

# Завдання 2: Пошук максимального елемента.Реверс списку.
Доповніть код, щоб знайти найбільший елемент у списку та вивести список в зворотньому порядку.

{44CFFD1B-43DA-4712-B211-6C3375F9C88D}.png

```python
n = int(input("Введіть кількість елементів: "))
mas = []

for i in range(n):
  num = int(input(f"Елемент {i+1}: "))
  mas.append(num)
max_element = mas[0]
for i in range(1,n):
  if mas[i] > max_element:
    max_element = mas[i]

reversed_mas = []
for i in range(n - 1, -1, -1):
  reversed_mas.append(mas[i])
print("Максимальний елемент:", max_element)
print("Список у зворотному порядку:", reversed_mas)
```

# Завдання 3: Сформувати список із
N випадкових цілих чисел у діапазоні від 1 до 100.
Обчислити суму всіх двоцифрових чисел у списку.

{5A18D894-3A51-4C35-8155-119408F7E5DB}.png

```python
import random

n = int(input("Введіть кількість елементів: "))

mas = []

for i in range(n):
  num = random.randint(1,100)
  mas.append(num)

s=0
for i in range(n):
  if 10 <= mas[i] <= 99:
    s+=mas[i]
print("Згенерований список:", mas)
print("Сума всіх двоцифрових чисел", s)
```

# Завдання 5: Створити масив, елементами якого є 10 випадкових чисел з діапазону -50;50 Обчислити добуток та кількість додатніх елементів масиву. Сортувати масив за зростанням метод екстремальних елементів.

{B2F0DD7B-7A8B-40A7-899D-EC2CEA2D81A3}.png

```python
import random
n=int(input('n='))
a=[]
for i in range(n):
    a.append(random.randint(-50,50))
print(a[i], end=" ")
print('\nМетод екстремальних елементів')
for j in range(n-1):
    min=j
    for i in range(j,n):
        if a[i]<a[min]:
            a[i],a[min]=a[min],a[i]
print(a)
d=1
k=0
for i in range(n):
    if i > 0 and a[i] != a[i-1]:
        d = d + 1
    if i == 0 or a[i] != a[i-1]:
        k = k + 1
print(f"d= {d}")
print(f"k= {k}")
```

# Завдання 6: Сформувати список з 10 випадкових цілих чисел від -100 до + 100.
Знайти мінімальний елемент списку і його порядковий номер. Отримати
інший список, що складається тільки з парних чисел вихідного списку або
повідомити, що таких чисел немає. Сортувати список за спаданням методом бульбашки.

{759965FE-A1B3-45C9-9F02-6F3FA48C474C}.png

```python
import random

n = int(input("Введіть кількість елементів: "))
mas = []

for i in range(n):
    num = random.randint(-100, 100)
    mas.append(num)

min_index = 0
for i in range(n):
    if mas[i] < mas[min_index]:
        min_index = i
min_element = mas[min_index]

parni_mas = []
for i in range(n):
    if mas[i] % 2 == 0:
        parni_mas.append(mas[i])


# Сортування методом бульбашки
for i in range(n - 1):
    for j in range(n - i - 1):
        if mas[j] < mas[j + 1]:
            mas[j], mas[j + 1] = mas[j + 1], mas[j]

print("Згенерований список:", mas)
print(f"Мінімальний елемент: {min_element}, Його індекс: {min_index}")
print("Парні числа:", parni_mas)
print("Відсортований список (за спаданням):", mas)
```
