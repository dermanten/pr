У Python двовимірні масиви (списки списків) працюють так само, як і класичні масиви в інших мовах програмування. Доступ до елементів здійснюється за допомогою індексів. Кожен елемент у масиві (який є списком) має свій індекс, і через ці індекси ми можемо отримувати або змінювати значення.

**Доступ до елементів здійснюється через два індекси:**

Перший індекс вказує на рядок.

Другий індекс — на стовпець у цьому рядку.

```python
for i in range(3):          # зовнішній цикл — рядки
    for j in range(4):      # внутрішній цикл — стовпці
        print(f"[{i},{j}]", end=' ')
    print()
```

# Обчислити суму двовимірного масиву, середнє значення.





```python
from random import randint


n = int(input("Кількість рядків (n): "))
m = int(input("Кількість стовпців (m): "))

# Створення і заповнення двовимірного масиву
mas = []

for i in range(n):
    row = []
    for j in range(m):
        row.append(randint(-50, 50))
    mas.append(row)

# Виведення масиву
print("\nМасив:")
for i in range(n):
    for j in range(m):
        print(f"{mas[i][j]:>4}", end=" ")
    print()

# Обчислення суми та середнього значення
s = 0
for i in range(n):
  for j in range(m):
    s +=mas[i][j]
ser = s/n*m

# Виведення результатів
print (f"Сума: {s}")
print(f"Середнє значення: {ser}")
```

# У двовимірному масиві визначити максимальний елемент та його індекси, поміняти місцями 1 та останній елемент.


```python
from random import randint

n = int(input("Кількість рядків (n): "))
m = int(input("Кількість стовпців (m): "))

a = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(randint(1, 100))
    a.append(row)

# Виведення масиву
print("\nПочатковий масив:")


# Пошук максимального
max_el = a[0][0]
max_i, max_j = 0, 0
for i in range(n):
    for j in range(m):
         if a[i][j] > max_el:
            max_el = a[i][j]
            max_i, max_j = i, j

print(f"\nМаксимальний елемент: {max_el}, індекси: [{max_i}][{max_j}]")

# Обмін першого та останнього елементів

a[0][0], a[n-1][m-1] = a[n-1][m-1], a[0][0]

# Виведення масиву
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:4}", end=" ")
    print()
```

# Піднести до квадрату всі парні значення елементів двовимірного масиву, знайти кількість парних елементів. Обчислити добуток парних елементів. Масив заповнити з клавіатури.


```python
n = int(input("Рядків: "))
m = int(input("Стовпців: "))

a = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(int(input(f"a[{i}][{j}] = ")))
    a.append(row)

count = 0
dob = 1
for i in range(n):
    for j in range(m):
      if a[i][j]%2==0:
        count += 1
        dob *= a[i][j]
        a[i][j] = a[i][j]**2

print("\nМасив після піднесення парних до квадрату:")
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:4}", end=" ")
    print()

print(f"\nКількість парних елементів: {count}")
print(f"Добуток парних елементів: {dob}")
```

# Піднести до квадрату всі від'ємні значення елементів двовимірного масиву, знайти кількість від'ємних елементів. Масив заповнити генератором випадкових чисел.

```python
from random import randint

n = int(input("Рядків: "))
m = int(input("Стовпців: "))

a = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(randint(-10, 10))
    a.append(row)

print("\nПочатковий масив:")
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:4}", end=" ")
    print()

count = 0
for i in range(n):
    for j in range(m):
        if a[i][j] < 0:
            a[i][j] = a[i][j]**2
            count += 1

print("\nМасив після обробки від’ємних:")
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:4}", end=" ")
    print()

print(f"\nКількість від’ємних елементів: {count}")
```

# Обчислити суму елементів кожного рядка та кожного стовпця

```python
from random import randint

# Введення розмірів масиву
rows = int(input("Кількість рядків: "))
cols = int(input("Кількість стовпців: "))

# Створення та заповнення масиву
a = []
for i in range(rows):
    row = []
    for j in range(cols):
        row.append(randint(1, 9))
    a.append(row)

# Виведення масиву
print("\nМасив:")
for i in range(rows):
    for j in range(cols):
        print(f"{a[i][j]:>4}", end=" ")
    print()

# Обчислення суми кожного стовпця
print("\nСума по стовпцях:")
for j in range(cols):
    col_sum = 0
    for i in range(rows):
        col_sum += a[i][j]
    print(f"Стовпець {j}: {col_sum}")

# Обчислення суми кожного рядка
print("\nСума по рядках:")
for i in range(rows):
    row_sum = 0
    for j in range(cols):
        row_sum += a[i][j]
    print(f"Рядок {i}: {row_sum}")
```

# Замінити всі парні числа в двовимірному масиві на 0, а всі непарні — на 1.

```python
from random import randint

n = int(input("Рядків: "))
m = int(input("Стовпців: "))

a = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(randint(1, 20))
    a.append(row)

print("\nПочатковий масив:")
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:4}", end=" ")
    print()

for i in range(n):
    for j in range(m):
      if a[i][j] % 2 ==0:
        a[i][j] = 0
      else:
        a[i][j] = 1


print("\nМасив після заміни парних на 0, непарних на 1:")
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:4}", end=" ")
    print()
```

# Бульбашкове сортування

Сортування парних стовпців  за зростанням, а для непарних — за спаданням.

```python
import random


n = int(input('n='))
m = int(input('m='))

# Створення та заповнення масиву
a = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(random.randint(1, 9))
    a.append(row)

# Виведення початкового масиву
print("Початковий масив:")
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:<4}", end=" ")
    print()

# Сортування стовпців
for j in range(m):  # Проходимо по стовпцях
    # Якщо стовпець непарний, сортуємо за спаданням
    if j % 2 == 1:
        for i in range(n - 1):
            for k in range(n - 1 - i):
                if a[k][j] < a[k + 1][j]:  # Якщо поточний елемент менший за наступний, міняємо їх
                    a[k][j], a[k + 1][j] = a[k + 1][j], a[k][j]
    else:  # Якщо стовпець парний, сортуємо за зростанням
        for i in range(n - 1):
            for k in range(n - 1 - i):
                if a[k][j] > a[k + 1][j]:  # Якщо поточний елемент більший за наступний, міняємо їх
                    a[k][j], a[k + 1][j] = a[k + 1][j], a[k][j]

# Виведення відсортованого масиву
print("\nПосортований масив:")
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:<4}", end=" ")
    print()
```

Сортування парних рядків  за зростанням, а для непарних — за спаданням.

```python
import random


n = int(input('n='))
m = int(input('m='))

# Створення та заповнення масиву
a = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(random.randint(1, 9))
    a.append(row)

# Виведення початкового масиву
print("Початковий масив:")
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:<4}", end=" ")
    print()

# Сортування рядків
for i in range(n):
    for j in range(m - 1):
        for k in range(m - 1 - j):
            if i % 2 == 0:
                # Парний рядок — сортувати за зростанням
                if a[i][k] > a[i][k + 1]:
                    a[i][k], a[i][k + 1] = a[i][k + 1], a[i][k]
            else:
                # Непарний рядок — сортувати за спаданням
                if a[i][k] < a[i][k + 1]:
                    a[i][k], a[i][k + 1] = a[i][k + 1], a[i][k]

# Виведення відсортованого масиву
print("\nПосортований масив:")
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:<4}", end=" ")
    print()
```
